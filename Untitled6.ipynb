```python
dataset = dataset.fillna(dataset.mean())
```

```python
print(dataset.isnull().sum())
```

```python
corr = dataset.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(dataset)
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Metodu')
plt.xlabel('Kume Sayisi')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data=principalComponents, columns=['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf, pd.DataFrame({'cluster': y_kmeans})], axis=1)
sns.heatmap(pd.DataFrame(pca.components_,columns=dataset.columns,index = ['PC-1','PC-2']), cmap='coolwarm')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data=principalComponents, columns=['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf, pd.DataFrame({'cluster': y_kmeans})], axis=1)
sns.scatterplot(data=finalDf, x='principal component 1', y='principal component 2', hue='cluster', palette='Dark2')
plt.show()
```
